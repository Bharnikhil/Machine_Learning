```python

from matplotlib.animation import FuncAnimation
import matplotlib.animation as animation
```

```python
def perceptron(X,y):
    
    m = []
    b = []
    
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1
    
    for i in range(1000):
        j = np.random.randint(0,200)
        y_hat = step(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]
        
        m.append(-(weights[1]/weights[2]))
        b.append(-(weights[0]/weights[2]))
        
    return m,b
```

```python
m,b = perceptron(X,y)
```

```python
fig, ax = plt.subplots(figsize=(9,5))

x_i = np.arange(-3, 4, 0.1)
y_i = x_i*m[0] +b[0]
ax.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
line, = ax.plot(x_i, x_i*m[0] +b[0] , 'r-', linewidth=2)
ax.set_xlim(-3, 4)
ax.set_ylim(-1.5, 4)
def update(i):
    label = 'epoch {0}'.format(i + 1)
    line.set_ydata(x_i*m[i] + b[i])
    ax.set_xlabel(label)
    # return line, ax

anim = FuncAnimation(fig, update, repeat=True, frames=1000, interval=100)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Fit logistic regression
lor = LogisticRegression()
lor.fit(X, y)
```

```python
# Extract coefficients and intercept
w0 = lor.coef_[0][0]  # coefficient for feature 1
w1 = lor.coef_[0][1]  # coefficient for feature 2
b = lor.intercept_[0] # bias/intercept

# Compute slope (m) and intercept (c) of the line
m_lor = -w0 / w1
c_lor = -b / w1
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Scatter plot of dataset
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', edgecolors='k')
plt.title("Synthetic Classification Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

# Plot the decision boundary
x_vals = np.array([X[:, 0].min() - 1, X[:, 0].max() + 1])
y_vals = m * x_vals + b
plt.plot(x_vals, y_vals, color='green', linewidth=2, label='Decision Boundary perceptron')

# Plot the lor boundary
x_vals = np.array([X[:, 0].min() - 1, X[:, 0].max() + 1])
y_vals = m_lor * x_vals + c_lor
plt.plot(x_vals, y_vals, color='black', linewidth=2, label='Decision Boundary lor')


# Set custom axis limits
plt.xlim(-1, 4)   # x-axis from -3 to 4
plt.ylim(-1.5, 4)   # y-axis from -1 to 4

plt.legend()
plt.show()

```
