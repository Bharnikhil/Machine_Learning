````python
# Load dataset
df = fetch_california_housing()

# Convert to DataFrame
dataset = pd.DataFrame(df.data, columns=df.feature_names)

````

````python
dataset.head()
````

````python
## Independent Features and dependent features 
X=dataset
y=df.target
````

````python
y
````

````python
## Train test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42)
````

````python
X_train
````

````python
from sklearn.linear_model import LinearRegression
````

````python
## Standardising the dataset
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
````

````python
X_train=scaler.fit_transform(X_train)
````

````python
X_test=scaler.transform(X_test)
````

##  Why `fit_transform` on X_train?

- **`fit()`**  Learns the transformation rule from the data  
  (e.g., mean & std for scaling, min & max for normalization).  

- **`transform()`**  Actually applies the learned transformation.  

So when we do:  

```python
X_train_scaled = scaler.fit_transform(X_train)


 Why only transform on X_test?
X_test_scaled = scaler.transform(X_test)


 Here, we apply the same rule learned from training to the test set.

This ensures the test data remains unseen during training, simulating real-world conditions.

 What if we use fit_transform on X_test also?

The scaler would calculate mean & std from the test set.

This causes data leakage  future/unseen data leaks into training.

The model will look artificially better during evaluation but fail in real-world use.

````python
X_train
````

````python
## to inverse the transformation applied 

scaler.inverse_transform(X_train)
````

````python
from sklearn.linear_model import LinearRegression
## cross validation

from sklearn.model_selection import cross_val_score
````

````python
regression=LinearRegression()
regression.fit(X_train,y_train)
````

````python
mse=cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',cv=5)
````

````python
np.mean(mse)
````

````python
## prediction
reg_pred=regression.predict(X_test)
````
